```python
from scipy import stats
import math
n = 200
p = 58/100
q = 42/100
np = n * p
nq = n * q
print(f'np :{np}, nq :{nq}')
# np and nq must be greater than 5 for use normally distributed

mu = np
sigma = math.sqrt(n * p * q)
x = 119.5
# x = continuity correction from 20 to 19.5
z_score = (x - mu) / sigma

value = 1 - stats.norm.cdf(z_score)
print(f'p: ',value)
# variable value is values under curve
```

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

def draw_z_score(x, cond, mu=0, sigma=1):
    y = stats.norm.pdf(x, mu, sigma)
    z = x[cond]
    plt.plot(x,y)
    plt.fill_between(z,0,stats.norm.pdf(z,mu,sigma))
    plt.show()
    
x = np.arange(-3,3,0.001)
draw_z_score(x, x>z_score)
```
